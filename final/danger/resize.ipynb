```python
import numpy as np
import os
import cv2

imgPath = './PNGImages_resize/'
newPath='./PNGImages_resizes/'

allFileList = os.listdir(imgPath)
for file in allFileList:
    if file==".DS_Store":
        continue
    file_name = file
    print(file)
    file1 = imgPath+file_name
    file2 =newPath+file_name#[0:len(file_name)-3]+"jpg"
    img = cv2.imread(file1)
    img = cv2.resize(img, (720,540))

    cv2.imwrite(file2, img)

```
